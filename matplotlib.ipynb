## **Part one**

**Basic Terminology**

###  **Key Terminology in Matplotlib**

#### 1. **Figure**

* The **overall window or image** that everything is drawn on.
* Think of it like a **canvas**.
* Figure jechuun iddoo wanti hunduu irratti kaafamu jechuudha.

```python
fig = plt.figure()
```


#### 2. **Axes**

* A **part** of the figure where **data is actually plotted**.
* Axes jechuun immoo part figure ta'ee kan kan irratti data'n kaafamu jechuudha. Figure'n tokko Axes baayyee of keessaa qabaachuu danda'a.
* One figure can have **multiple Axes (plots)**.
* It's what we usually call a **plot or graph**.

```python
ax = fig.add_subplot(1, 1, 1)  # One row, one column, first plot
```

#### 3. **Axis**

* The **x-axis** , **y-axis** and **z-axis** of a plot (inside the Axes).
* Axis'n immoo Axes irratti jechuudha. x-axis fi y-axis ibsu jechuudha.
* They define the **scale** and **ticks** (e.g., 0, 1, 2,...).

#### 4. **Plot**

* The **actual graph or line/bar/point** you see.
* Example: Line graph, bar chart, scatter plot.

```python
plt.plot([1, 2, 3], [4, 5, 6])  # a line plot
```

#### 5. **Label**

* Names for the x-axis, y-axis, or lines.

```python
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("My Plot")
```

#### 6. **Legend**

* A **box that explains what each line or color means** in your plot.
* legend jechuun immoo functionality chart nuti kaasne sanaa kan ibsu jechuudha.
```python
plt.legend(["Line 1"])
```


#### 7. Marker:

A **marker** is a **symbol** that shows each **data point** on a plot.
* jechuun kanneen akka circle xixiqqoo ykn star fi kkkf kanneen nuti fakkii irratti kaasnu jechuudha.
* isaan kanneen immmoo customize gochuu ni dandeenya. fknf color isaanii jijjiiruu fi kkf jechuudha.

Example:

```python
plt.plot(x, y, marker='o')
```

This adds a **circle** at each data point.


####  Common Markers:

* `'o'`  Circle
* `'x'`  X
* `'s'`  Square
* `'^'`  Triangle Up
* `'*'`  Star


#### 8. **Ticks**

* The **values shown on the axes** (e.g., 0, 1, 2...).
* You can customize them:

```python
plt.xticks([0, 1, 2])
plt.yticks([10, 20, 30])
```

#### 9. **Grid**

* Adds **grid lines** behind the plot for easier reading.

```python
plt.grid(True)
```

#### 10. **Style**

* The **look and feel** of the plot (color, line style, marker type).

```python
plt.plot(x, y, color='green', linestyle='--', marker='o')
```

#### 11. **Subplot**

* Multiple plots **inside one figure**.

```python
plt.subplot(2, 1, 1)  # 2 rows, 1 column, 1st plot
```

### What is lines in matplotlib?
* A **line** in Matplotlib is a** straight or curved** path that connects a series of data points on a 2D plottypically used in line graphs.

We can change using:

### What is `linewidth`?

In Matplotlib, `linewidth` (or `lw`) controls the **thickness** of the line drawn in a plot.


#### Syntax

```python
plt.plot(x, y, linewidth=2)
# or
plt.plot(x, y, lw=2)
```

Both `linewidth` and `lw` mean the same.


#### Default Value

* The default line width is `1.5`.

---

#### Common Values and Effects:

| `linewidth` Value | Description         |
| ----------------- | ------------------- |
| `0.5`             | Very thin line      |
| `1`               | Thin line (default) |
| `2`               | Medium thickness    |
| `3` or more       | Thick line          |

---

### Example:


````python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [2, 3, 5, 7]

plt.plot(x, y, linewidth=3)
plt.title("Line with Custom Width")
plt.show()
````

#### Combine with Other Styles:

You can use `linewidth` with color, linestyle, and marker:


**Grid function**

**Grid lines** are horizontal and vertical lines across the plot that help you `visually align data points` with the axes. They're useful for reading values accurately.

## **Part Seven**

### **Subplots in matplotlib**

####  What is a **Subplot**?

A **subplot** is a way to show **multiple plots in one figure**organized in rows and columns.


#### Basic Syntax

```python
plt.subplot(nrows, ncols, index)
```

* `nrows`: number of rows
* `ncols`: number of columns
* `index`: the position of the current subplot (starts at 1)


#### Example:

````python
import matplotlib.pyplot as plt

x = [1, 2, 3]
y1 = [2, 4, 1]
y2 = [1, 3, 5]

plt.subplot(2, 1, 1)  # 2 rows, 1 column, 1st plot
plt.plot(x, y1)
plt.title("First Plot")

plt.subplot(2, 1, 2)  # 2nd plot
plt.plot(x, y2)
plt.title("Second Plot")

plt.tight_layout()
plt.show()
````



#### Tips:

* Use `plt.tight_layout()` to fix overlapping titles/labels.
* You can create grids like:

  * `2x2` (4 plots)
  * `1x3` (3 plots in a row)
  * `3x1` (3 plots in a column)


#### Advanced: `plt.subplots()`

This gives more control and returns both **Figure** and **Axes** objects:

```python
fig, axs = plt.subplots(1, 2)  # 1 row, 2 columns
axs[0].plot(x, y1)
axs[1].plot(x, y2)
```

## **`legend()`** function


#### `legend()` in Matplotlib

The `legend()` function in Matplotlib is used to **display a legend box**, which tells you **what each line, marker, or bar represents** in a plot.


####  **Basic Usage**

````python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
y2 = [2, 3, 5, 10]

plt.plot(x, y1, label="Line A")  # label used for legend
plt.plot(x, y2, label="Line B")

plt.legend()  # Show legend
plt.show()
````



## **How it Works**

| Parameter      | Description                                                                      |
| -------------- | -------------------------------------------------------------------------------- |
| `label`        | Used inside `plot()` or any chart function to define what shows up in the legend |
| `plt.legend()` | Displays the legend box                                                          |
| `loc`          | Position of the legend                                                           |
| `fontsize`     | Size of the text in legend                                                       |
| `title`        | Title for the legend box                                                         |
| `frameon`      | Show/hide the legend box border (True/False)                                     |
| `ncol`         | Number of columns (for horizontal legends)                                       |



#### **Common Parameters of `legend()`**


````python
plt.legend(
    loc="upper right",  # Position
    fontsize=12,        # Font size
    title="Legend Title",  # Optional title
    frameon=True,       # Show border (default True)
    ncol=2              # 2 columns (for horizontal style)
)
````


#### **Legend Location Options (`loc`)**

| Location Name    | Description             |
| ---------------- | ----------------------- |
| `'best'`         | Auto location (default) |
| `'upper right'`  | Top-right corner        |
| `'upper left'`   | Top-left corner         |
| `'lower left'`   | Bottom-left corner      |
| `'lower right'`  | Bottom-right corner     |
| `'center'`       | Center of the plot      |
| `'center left'`  | Left-center             |
| `'center right'` | Right-center            |

> Example: `plt.legend(loc='upper left')`

#### **Customizing the Legend Box**


````python
plt.legend(
    loc="upper right",
    fontsize=10,
    title="Lines",
    frameon=True,
    facecolor="lightgray",  # background color
    edgecolor="black"       # border color
)
````

#### **Legend Without Labels?**

If you forgot to set `label=` in `plot()`, legend will show nothing. You must label like this:

````python
plt.plot(x, y1, label="My Line")
plt.legend()
````

#### Example with All Customizations


````python
x = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
y2 = [2, 3, 5, 10]

plt.plot(x, y1, label="Square", color="blue")
plt.plot(x, y2, label="Linear", color="green")

plt.legend(loc="lower right", fontsize=12, title="Functions", frameon=True, facecolor="lightyellow")
plt.title("Legend Example")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.grid(True)
plt.show()
````

#### What is `facecolor`?

`facecolor` sets the **background color** of a Matplotlib object, such as:

* The **legend box**
* A **figure** or **axes**
* A **bar** or **patch**
* A **plot area**

Scatter plots help you:

## **Grouped Bar Plot**

Let's now create a **Grouped Bar Plot**  also called a **side-by-side bar chart**  great for comparing multiple values **across the same categories** (e.g., 2024 vs 2025 sales).


###  What Is a Grouped Bar Plot?

It shows **two or more bars per category**, grouped side by side to:

* Compare **different years**, **genders**, **regions**, etc.
* Highlight differences **within each group**.


####  **Implementation Example: 2024 vs 2025 Fruit Sales**



````python
import matplotlib.pyplot as plt
import numpy as np

# Categories (X-axis)
categories = ['Apples', 'Bananas', 'Cherries', 'Dates']
x = np.arange(len(categories))  # positions for bars

# Values for two years
sales_2024 = [20, 35, 30, 10]
sales_2025 = [25, 40, 33, 15]

# Bar width
width = 0.4

# Create side-by-side bars
plt.bar(x - width/2, sales_2024, width=width, label='2024', color='skyblue')
plt.bar(x + width/2, sales_2025, width=width, label='2025', color='orange')

# Labeling and legend
plt.xlabel("Fruit Type")
plt.ylabel("Quantity Sold")
plt.title("Fruit Sales Comparison: 2024 vs 2025")
plt.xticks(x, categories)  # set category labels on x-axis
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.5)

# Show plot
plt.show()
````


#### Key Concepts in Code:

| Concept           | Explanation                       |
| ----------------- | --------------------------------- |
| `x = np.arange()` | Create positions for categories   |
| `width`           | Width of each bar                 |
| `x - width/2`     | Shift 1st group bars to the left  |
| `x + width/2`     | Shift 2nd group bars to the right |
| `plt.xticks()`    | Set category names on the x-axis  |

###  **Stacked Bar Plot in Matplotlib**

 Let's move on to the **Stacked Bar Plot**  ideal for showing how **parts make up a whole** in each category.

#### What is a Stacked Bar Plot?

A **stacked bar plot** places bars **on top of each other** to represent **sub-groups** within a category.
Instead of placing them side-by-side (like grouped bars), each segment **adds to the total height**.


#### Use Case:

* Visualizing **total fruit sales** with breakdowns by year (e.g., 2024 stacked over 2025).
* Showing **male and female counts** per department.
* Comparing **expenses** in different categories over time.


#### **Implementation Example: Stacked Fruit Sales**

````python
import matplotlib.pyplot as plt
import numpy as np

# Categories
categories = ['Apples', 'Bananas', 'Cherries', 'Dates']
x = np.arange(len(categories))

# Sales data
sales_2024 = [20, 35, 30, 10]
sales_2025 = [25, 40, 33, 15]

# Plot stacked bars
plt.bar(x, sales_2024, label='2024', color='lightblue')
plt.bar(x, sales_2025, bottom=sales_2024, label='2025', color='orange')

# Labels and title
plt.xlabel("Fruit Type")
plt.ylabel("Total Sales")
plt.title("Stacked Fruit Sales: 2024 & 2025")
plt.xticks(x, categories)
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()
````

| Code                  | Description                                          |

| `bottom=sales_2024`   | Tells matplotlib where to start the second bar stack |
| `color=` and `label=` | Used for coloring and legend                         |

## **Pie Chart**


#### What Is a Pie Chart?

A **pie chart** is a circular chart divided into **slices**, where each slice represents a **categorys proportion** of the total.
Used for **comparing parts of a whole** at a glance.

<font color = 'green'> A **pie chart** is a circular chart divided into sectors, `each` representing a **proportion** of the whole. It's used `to `**compare parts of a whole**.
</font>

>A pie chart is a circular statistical graphic divided into slices, where:

    * Each slice represents a category
    * The size of the slice represents the proportion or percentage of the total
    * The whole circle represents 100% (the full dataset)

#### When to Use:

* To show how a **budget is divided**
* Market share by brand
* Survey results (e.g., device usage)
* To show parts of a whole
* we have 5 of fewer categories
* To compare properties visually

<font color='green'>

**Generally:**

  * To show percentage breakdowns (e.g., market share, budget distribution).
  * When you have categorical data and want to show their relative sizes.
</font>

##### **Example Situations**

| Situation                         | Reason to Use Pie Chart           |
| --------------------------------- | --------------------------------- |
| Device usage: Android, iOS, Other | Show what % uses each device type |
| Budget: Rent, Food, Transport     | Show how money is divided         |
| Survey results                    | Show responses distribution       |

### **Important Pie Chart Terms**
| Term           | Description                                         |
| -------------- | --------------------------------------------------- |
| **Slice**      | A wedge in the pie (a category)                     |
| **Labels**     | Category names shown on slices                      |
| **Size/Value** | The actual number/percentage the slice represents   |
| **autopct**    | Shows % text on each slice                          |
| **startangle** | Rotates where the pie starts                        |
| **explode**    | Pulls a slice out to highlight it                   |
| **colors**     | Custom color for each slice                         |
| **shadow**     | Adds a shadow for depth effect                      |
| **equal axis** | Makes the pie perfectly round (`plt.axis('equal')`) |


When Not to Use a Pie Chart:
* If you have too many categories (bar chart is better)

* If you want to show trends over time (use line chart)

* If proportions are very similar (hard to visually compare)

You have a monthly budget for your expenses:

* Rent: 40%
* Food: 25%
* Transport: 15%
* Entertainment: 10%
* Savings: 10%

Plot this data as a pie chart with:

* Labels
* Percentages
* An exploded slice for **Rent**
* Title: `"Monthly Budget Breakdown"`
* Make it circular (`axis('equal')`)

### Starter Code:

````python
import matplotlib.pyplot as plt

labels = ['Rent', 'Food', 'Transport', 'Entertainment', 'Savings']
sizes = [40, 25, 15, 10, 10]
explode = [0.1, 0, 0, 0, 0]
colors = ['lightblue', 'lightgreen', 'orange', 'violet', 'grey']

plt.pie(sizes, labels=labels, explode=explode, colors=colors,
        autopct='%1.1f%%', startangle=140, shadow=True)

plt.title("Monthly Budget Breakdown")
plt.axis('equal')
plt.show()
````


### <font color="blue">**Donut Chart (Pie Chart with a Hole)** </font>

A **donut chart** is just a pie chart with a circle in the middle.

#### Example:

````python
plt.pie(sizes, labels=labels, colors=colors,
        autopct='%1.1f%%', startangle=140)

# Add a white circle in the center
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title("Monthly Budget as Donut Chart")
plt.axis('equal')
plt.show()
````

### **Histograms**

#### What is a Histogram?

A **histogram** is a type of bar plot that shows the **frequency distribution** of numerical data. It answers questions like:

> How many data points fall into this `range ` (bin)? <font color ="blue"> `bin` means`range`. kanaaf bin= 5 jechuun range shanitti qoqqoodna jechuudha. </font>

> Kana jechuun gaaffilee kanneen akka data point meeqatu range kanaaf kana gidduu jiraa isa jedhu addaan baafachuufaa jechuudha. <font color="green"> Fakkeenyaaf barataan sadii 10 fi 20 gidduu barataan shan 20 fi 30 gidduu barataan afur 30 fi 40 gidduu hojjete jennee kan qoqqoodnu yoo ta'e jechuudha</font>

#### When to Use a Histogram:

* To understand the **distribution** of your data.
* To identify **patterns**, such as skewness, peaks, or outliers.
* To **group continuous data** into intervals (called **bins**).


#### Key Terminology

| Term          | Meaning                                             |
| ------------- | --------------------------------------------------- |
| **Bins**      | Intervals or ranges the data is grouped into.       |
| **Frequency** | Number of values that fall into each bin.           |
| **Density**   | Shows probability density (used in `density=True`). |


#### Basic Syntax for Histogram

```python
plt.hist(data, bins=number_of_bins)
```

#### Simple Example


````python
import matplotlib.pyplot as plt

# Sample data
data = [10, 12, 13, 15, 15, 17, 19, 20, 20, 20, 22, 24, 25, 25, 30]

# Create histogram
plt.hist(data, bins=5, color='skyblue', edgecolor='blue')
# bins= 5 jechuun bakka shanitti qoqqoode jechuudha. Range shanitti qoqqoode jechuudha.

# Add titles and labels
plt.title("Simple Histogram")
plt.xlabel("Value Range")
plt.ylabel("Frequency")

# Show plot
plt.show()
````

#### How It Works:

* `data`: List of numbers (continuous values).
* `bins=5`: Groups the data into 5 ranges.
* `color`: Fill color of the bars.
* `edgecolor='black'`: Adds borders to the bars for clarity.


#### Optional Parameters

| Parameter      | Use                                 |
| -------------- | ----------------------------------- |
| `bins`         | Number of bins or list of bin edges |
| `color`        | Bar color                           |
| `edgecolor`    | Border color of bars                |
| `density=True` | Shows density instead of frequency  |
| `alpha`        | Transparency of bars (0 to 1)       |



````python
# tokkoon tokkoon isaa width gidduu jiru addaaan baasuu dandeenya jechuudha.
# rwidth fayyadamna jechuudha. kana jechuun row width jechuudha.
import matplotlib.pyplot as plt

# Sample data
data = [10, 12, 13, 15, 15, 17, 19, 20, 20, 20, 22, 24, 25, 25, 30]

# Create histogram
plt.hist(data, bins=5, color='skyblue', edgecolor='blue', rwidth=0.7)
# bins= 5 jechuun bakka shanitti qoqqoode jechuudha. Range shanitti qoqqoode jechuudha.

# Add titles and labels
plt.title("Simple Histogram")
plt.xlabel("Value Range")
plt.ylabel("Frequency")

# Show plot
plt.show()
````

#### Advanced Example: Random Data


````python
import matplotlib.pyplot as plt
import numpy as np

# Random data from normal distribution
data = np.random.randn(1000)

# Plot histogram
plt.hist(data, bins=30, color='purple', edgecolor='black' , alpha=0.7)

plt.title("Histogram of Normally Distributed Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
````
